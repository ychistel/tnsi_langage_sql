```sql
CREATE TABLE IF NOT EXISTS Auteurs (
    id_auteur INTEGER NOT NULL PRIMARY KEY,
    nom VARCHAR(70) NOT NULL,
    prénom VARCHAR(70) NOT NULL
);

CREATE TABLE IF NOT EXISTS Livres (
    id_livre INTEGER NOT NULL PRIMARY KEY,
    titre VARCHAR(50) NOT NULL,
    prixHT INT,
    année INT,
    id_genre INT REFERENCES Genres(id_genre),
    id_editeur INT REFERENCES Editeurs(id_editeur)
);

CREATE TABLE IF NOT EXISTS Editeurs (
    id_editeur INTEGER NOT NULL PRIMARY KEY,
    nom VARCHAR(30) NOT NULL
);

CREATE TABLE IF NOT EXISTS Genres (
    id_genre INTEGER NOT NULL PRIMARY KEY,
    genre VARCHAR(50) NOT NULL
);

CREATE TABLE IF NOT EXISTS Ecrits (
    id_auteur INT REFERENCES Auteurs(id_auteur),
    id_titre INT REFERENCES Livres(id_livre)
);
```

## Requetes SQL

a) Les titres des livres de la base.

```sql
SELECT * FRoM livres;
```

```sql
select * from auteurs;
```

b) Les titres et années de parution des livres.

```sql
SELECT titre, année FROM livres;
```

c) Les titres des livres qui commencent par la lettre « T ».

```sql
SELECT titre FROM livres WHERE titre LIKE 't%';
```

d) Le prix HT maximal d’un livre de la base.

```sql
SELECT max(prixHT) FROM livres;
```

e) Tous les champs de la table Auteurs.

```sql
SELECT * FROM auteurs;
```

```sql
UPDATE auteurs SET prénom = 'Christopher', nom = 'Nolan' WHERE id_auteur = 15
```

f) Le nombre d’auteurs contenus dans la base.

```sql
SELECT COUNT(nom) FROM auteurs;
```

g) Le nombre de livres référencés dans la base et le prix moyen.

```sql
SELECT COUNT(id_livre),AVG(prixHT),SUM(prixHT),SUM(prixHT)/COUNT(id_livre) FROM livres;
```

h) Les noms et prénoms de tous les auteurs par ordre alphabétique de nom.

```sql
SELECT nom,prénom FROM auteurs
ORDER BY nom ASC;
```

i) Les noms des auteurs dont le prénom est « Pierre ».

```sql
SELECT nom FROM auteurs WHERE prénom = 'Pierre';
```

j) Les titres des livres coûtant plus de 15 euros HT avec leur prix HT.

```sql
SELECT titre,prixHT FROM livres WHERE prixHT > 15;
```

k) Les titres de livres coûtant moins de 15 euros avec leur prix TTC. La TVA est de 5,5%.

```sql
SELECT titre,prixHT*1.055 FROM livres WHERE prixHT < 15;
```

l) Les titres et années de parution des livres parus de 2010 à 2015 ordonnés suivant l’année de parution de manière décroissante.

```sql
SELECT titre,année FROM livres WHERE année BETWEEN 2010 and 2015 ORDER BY année DESC;
```

m) Les titres des livres avec le nom de leur éditeur.

```sql
SELECT titre,nom FROM livres
JOIN editeurs ON livres.id_editeur = editeurs.id_editeur
```

n) Les titres des livres avec le genre « science ».

```sql
SELECT titre, genre FROM livres
JOIN genres ON livres.id_genre = genres.id_genre
WHERE genre = 'science';
```

o) Les titres et les prix HT des livres du genre « policier » coûtant moins de 20 euros HT.

```sql
SELECT titre, prixHT FROM livres
JOIN genres ON livres.id_genre = genres.id_genre
WHERE genre = 'policier' and prixHT < 20;
```

p) Les années de parution de livres dans le genre « science » sans doublons.

```sql
SELECT DISTINCT année, titre FROM livres
JOIN genres ON livres.id_genre = genres.id_genre
WHERE genre = 'science';
```

q) Le prix total de tous les livres parus en 2019 dans le genre « science ».

```sql
SELECT SUM(prixHT) FROM livres
JOIN genres ON livres.id_genre = genres.id_genre
WHERE genre = 'science' and année = 2019;
```

r) Les identifiants des livres dont l’auteure a pour prénom « Marie ».

```sql
SELECT id_livre FROM livres
JOIN ecrits ON livres.id_livre = ecrits.id_titre
JOIN auteurs ON ecrits.id_auteur = auteurs.id_auteur
WHERE prénom = 'Marie';
```

s) Les titres des livres parus avant 2019 contenant la chaine « pil » dans le titre pour le genre 'histoire'.

```sql
SELECT titre,nom FROM livres
JOIN genres ON livres.id_genre = genres.id_genre
JOIN ecrits ON livres.id_livre = ecrits.id_titre
JOIN auteurs ON auteurs.id_auteur = ecrits.id_auteur
WHERE année = 2019 AND titre LIKE '%nsi%' and genre = 'science';
```

t) Le titre du livre (ou les titres des livres) dont le prix est maximal.

```sql
SELECT titre,prixHT FROM livres
WHERE prixHT = (SELECT max(prixHT) FROM livres);
```

u) Le titre du livre (ou les titres des livres) dont le prix est maximal en 2015.

```sql
SELECT titre,prixHT,année FROM livres
WHERE prixHT = (select max(prixHT) from livres WHERE année = 2015);
```

v) Les titres des livres dont le prix est plus petit que le prix moyen avec leur prix.

```sql
SELECT titre, prixHT FROM livres
WHERE prixHT < (SELECT AVG(prixHT) from livres);
```

w) Les titres du livre le plus cher et du livre le moins cher avec leur prix. Il peut y en avoir plusieurs.

```sql
SELECt titre,prixHT FROM livres
WHERE prixHT = (SELECT MAX(prixHT) FROM livres) OR prixHT = (SELECT MIN(prixHT) FROM livres);
```

x) Les titres des livres coûtant le plus cher avec le nom de leurs éditeurs.

```sql
SELECT titre,nom FROM livres
JOIN editeurs ON livres.id_editeur = editeurs.id_editeur
WHERE prixHT = (SELECT MAX(prixHT) FROM livres);
```

y) Les genres pour lesquels il n’y a aucun livre.

```sql
SELECT id_genre,genre FROM genres
WHERE id_genre NOT IN (SELECT id_genre FROM livres);
```
